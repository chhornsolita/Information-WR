```python
from gensim.models import FastText
from nltk.tokenize import word_tokenize

#Try dataset
sentences = ["FastText embedding handle subword information.",
             "It is effective for varius Languages.",]

# Tokenize sentences
tokenized_sentences = [word_tokenize(sentence.lower()) for sentence in sentences]

# Train FastText model
model = FastText(sentences=tokenized_sentences, vector_size=100, window=5, min_count=1, workers=4)

# Access word embeddings
word_embedding = model.wv
print(word_embedding['subword'])
```
